```python
vax_2 = pd.read_csv('data/county_vax.csv')
vax_2['administration_date'] = pd.to_datetime(vax_2.administration_date)
vax_2
```

```python
one_dose = vax_2[vax_2.dose_number ==1].groupby(['fips','locality','Population']).vaccine_doses_administered.sum().reset_index()
one_dose.columns = ['fips','locality','population','One_dose']
one_dose
```

```python
full_vax = vax_2[(vax_2.dose_number ==2)|((vax_2.dose_number ==1) & (vax_2.vaccine_manufacturer =='J&J'))].groupby(['fips','locality','Population']).vaccine_doses_administered.sum().reset_index()
full_vax.columns = ['fips','locality','population','Fully_vaccinated']
full_vax
```

```python
total_dose = vax_2.groupby(['fips','locality','Population']).vaccine_doses_administered.sum().reset_index()
total_dose.columns = ['fips','locality','population','Total_doses']
total_dose
```

```python
from functools import reduce
data_frames = [total_dose, one_dose, full_vax]
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['fips','locality','population'],
                                            how='left'), data_frames)
```

```python
df_merged
```

```python
df_merged['Total_doses_rate'] = round(df_merged.Total_doses / df_merged.population * 100000).astype(int)
df_merged['One_doses_rate'] = round(df_merged.One_dose / df_merged.population * 100000).astype(int)
df_merged['Fully_vaccinated_rate'] = round(df_merged.Total_doses / df_merged.population * 100000).astype(int)
```

```python
df_merged
```

```python
df_merged.info()
```

```python
county_df = vax_2.copy()
```

```python
one_dose = county_df[county_df.dose_number ==1].groupby(['fips','locality','Population']).vaccine_doses_administered.sum().reset_index()
one_dose.columns = ['fips','locality','population','One_dose']
full_vax = county_df[(county_df.dose_number ==2)|((county_df.dose_number ==1) & (county_df.vaccine_manufacturer =='J&J'))].groupby(['fips','locality','Population']).vaccine_doses_administered.sum().reset_index()
full_vax.columns = ['fips','locality','population','Fully_vaccinated']
total_dose = county_df.groupby(['fips','locality','Population']).vaccine_doses_administered.sum().reset_index()
total_dose.columns = ['fips','locality','population','Total_doses']

#merge dfs
data_frames = [total_dose, one_dose, full_vax]
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['fips','locality','population'],
                                            how='left'), data_frames)
```

```python
fig = px.choropleth(df_merged, geojson=counties,
                        locations='fips',
                        color='Fully_vaccinated',
                        range_color=(0,df_merged['Fully_vaccinated'].max()*0.5),
                        color_continuous_scale="mint",
                        scope='usa',
                        hover_name='locality',
                        hover_data=['Total_doses','One_dose','Fully_vaccinated'])
fig.update_geos(fitbounds="locations",visible=False)
fig.update_layout(margin={"r":0,"l":0,"b":0})
plot(fig)
```
